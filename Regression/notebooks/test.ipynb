```python
import joblib
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
```

# XGBoost

## Załadowanie modeli

```python
xgb_cv_model = joblib.load('saved_models/XGB_CV_Model.pkl')
xgb_gs_model = joblib.load('saved_models/XGB_GS_Model.pkl')
xgb_rgs_model = joblib.load('saved_models/XGB_RGS_Model.pkl')
```

## Załadowanie zbioru testowego

```python
df = pd.read_csv('saved_models/XGB_test_set.csv')
X_test = df.drop('target', axis=1)
y_test = df['target']
```

## Model - kroswalidacja

```python
cv_preds = xgb_cv_model.predict(X_test)

cv_mse = mean_squared_error(y_test, cv_preds)
cv_rmse = np.sqrt(cv_mse)
cv_r_squared = r2_score(y_test, cv_preds)

print("CV_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(cv_mse, cv_rmse, cv_r_squared))
```

## Model - Grid Search

```python
gs_preds = xgb_gs_model.predict(X_test)

gs_mse = mean_squared_error(y_test, gs_preds)
gs_rmse = np.sqrt(gs_mse)
gs_r_squared = r2_score(y_test, gs_preds)

print("GS_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(gs_mse, gs_rmse, gs_r_squared))
```

## Model - Randomized Grid Search

```python
rgs_preds = xgb_rgs_model.predict(X_test)

rgs_mse = mean_squared_error(y_test, rgs_preds)
rgs_rmse = np.sqrt(rgs_mse)
rgs_r_squared = r2_score(y_test, rgs_preds)

print("RGS_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(rgs_mse, rgs_rmse, rgs_r_squared))
```

# LinearRegression

```python
lr_cv_model = joblib.load('saved_models/LR_CV_Model.pkl')

df = pd.read_csv('saved_models/LR_test_set.csv')
X_test = df.drop('target', axis=1)
y_test = df['target']

lr_preds = lr_cv_model.predict(X_test)
mse = mean_squared_error(y_test, lr_preds)
rmse = np.sqrt(mse)
r_squared = r2_score(y_test, lr_preds)

print("LR_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(mse, rmse, r_squared))
```

# Random forest

```python
rf_cv_model = joblib.load('saved_models/RF_CV_Model.pkl')

df = pd.read_csv('saved_models/RF_test_set.csv')
X_test = df.drop('target', axis=1)
y_test = df['target']

rf_preds = rf_cv_model.predict(X_test)
mse = mean_squared_error(y_test, rf_preds)
rmse = np.sqrt(mse)
r_squared = r2_score(y_test, rf_preds)

print("LR_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(mse, rmse, r_squared))
```
