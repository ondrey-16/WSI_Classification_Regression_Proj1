```python
import joblib
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
```

# Załadowanie modeli

```python
cv_model = joblib.load('saved_models/CV_Model.pkl')
gs_model = joblib.load('saved_models/GS_Model.pkl')
rgs_model = joblib.load('saved_models/RGS_Model.pkl')
```

# Załadowanie zbioru testowego

```python
df = pd.read_csv('saved_models/test_set.csv')
X_test = df.drop('target', axis=1)
y_test = df['target']
```

# Model - kroswalidacja + prosty trening

```python
cv_preds = cv_model.predict(X_test)

cv_mse = mean_squared_error(y_test, cv_preds)
cv_rmse = np.sqrt(cv_mse)
cv_r_squared = r2_score(y_test, cv_preds)

print("CV_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(cv_mse, cv_rmse, cv_r_squared))
```

# Model - Grid Search

```python
gs_preds = gs_model.predict(X_test)

gs_mse = mean_squared_error(y_test, gs_preds)
gs_rmse = np.sqrt(gs_mse)
gs_r_squared = r2_score(y_test, gs_preds)

print("GS_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(gs_mse, gs_rmse, gs_r_squared))
```

# Model - Randomized Grid Search

```python
rgs_preds = rgs_model.predict(X_test)

rgs_mse = mean_squared_error(y_test, rgs_preds)
rgs_rmse = np.sqrt(gs_mse)
rgs_r_squared = r2_score(y_test, rgs_preds)

print("RGS_Model test: MSE = {}, RMSE = {}, R^2 = {}".format(rgs_mse, rgs_rmse, rgs_r_squared))
```
